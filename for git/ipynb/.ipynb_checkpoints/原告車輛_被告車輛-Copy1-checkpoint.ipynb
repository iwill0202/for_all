```python
vehicle = {
u'腳踏車':[u'腳踏自行車'],
u'三輪車':[u'三輪腳踏車',u'農用三輪車',u'三輪車',u'拼裝動力三輪車',u'拼裝三輪車',u'機器三輪車',u'馬達三輪車',u'摩拖三輪車',u'電動三輪車',u'動力三輪車',u'單速三輪車',u'大型三輪車',u'醫療用電三輪車',u'電動代步車',u'拼裝附動力三輪車',u'殘障三輪車',u'身心障礙者生活輔助器具特製車',u'殘障動力三輪車'],
u'小客車':[u'不明車號汽車',u'不明車輛',u'不明汽車',u'自小客',u'自小客車',u'自用小客車',u'租賃小客車',u'租賃用小客車',u'休旅車'],
u'營業小客車':[u'計程車',u'營業小客車',u'營業用小客車'],
u'輕型機車':[u'機器腳踏車',u'普通輕型機車',u'電動機車'],
u'重型機車':[u'警用重型機車',u'重型機車',u'普通重機車',u'普通重型機車',u'電動重型機車'],
u'大客車':[u'遊覽車',u'出租遊覽車',u'大客車',u'營業大客車',u'公車',u'客運'],
u'農用車':[u'農耕機',u'農耕曳引機',u'農耕曳引車',u'農耕用車'],
u'火車':[u'火車',u'電力機車'],
u'卡車':[u'曳引車',u'全拖車',u'貨櫃車',u'聯結車',u'大卡車',u'貨車'],
u'拖板車':[u'拖板車',u'電動拖板車']
}
#u'特種車':[u'油罐車',u'垃圾車',u'灑水車',u'吊車',u'救濟車',u'消防車',u'救護車',u'警備車',u'憲警巡邏車',u'工程車',u'教練車',u'殘障用特製車',u'郵車',u'清掃車',u'水肥車',u'囚車',u'殯儀館運靈車',u'靈車',u'油罐車',u'灑水車',u'垃圾車',u'運鈔車',u'救護車']

def find_vehicle(ls):# TT:what's "ls"?
    Plaintiffs_vehicle = []#TT:開空間放結果
    Defendants_vehicle = []
    for i in range(len(ls)):#------------------------1st for
                                #ls[i]
        for v in vehicle.keys():#------------------2nd for "TT:find the keys in the dictionary.'.key()' returns a list"
            
            for c in vehicle[v]:#------------------3rd for "TT:'c' is any key of vehicle key list."
                if c in ls[i]:# TT:what's "i"? get the "i"th item of ls list?

                    if u'原告' not in ls[i] and u'被告' not in ls[i]:#若是兩個都沒有(TT:在ls[i]中，)往前找
                        for minus in range(1,3):
                            if u'原告' in ls[i-minus] and u'被告' in ls[i-minus]:#TT:ls list到底怎麼切的?@@
                                if ls[i-minus].find(u'原告')< ls[i-minus].find(u'被告'):
                                #TT:[同一句?該段落?中順序是[原告]->[被告]->[交通工具]，則該交通工具是被告的]?
                                
#                                     print '----'
                                    
                                    Defendants_vehicle.append(v)#TT:append v，把被告交通工具加上找到的key name
                                    break
                                
                                else:
#                                     print u'原告',v,c,ls[i-minus],ls[i]
#                                     print '----'
                                    
                                    Plaintiffs_vehicle.append(v)#TT:如果不是"[原告]->[被告]->[交通工具]"這個順序，就是原告的車
                                    break
                            
                            elif u'原告' in ls[i-minus]:#TT:如果在minus中只找到原告，就是原告的車
#                                 print u'原告',v,c,ls[i-minus]
#                                 print '----'
                                
                                Plaintiffs_vehicle.append(v)
                                break
                            
                            elif u'被告' in ls[i-minus]:#TT:如果在minus中只找到被告
#                                 print u'被告',v,c,ls[i-minus]
#                                 print '----'
                                
                                Defendants_vehicle.append(v)
                                break
                                
                    elif u'原告' in ls[i] and u'被告' in ls[i]:#TT:如果找到原告與被告在 ls[i]中
                        if ls[i].find(u'被告') < ls[i].find(u'原告') < ls[i].find(c):
                            #TT:[如果順序是[被告]->[原告]->[交通工具]，則該交通工具是原告的](交通工具的詞離原告比較近)?

#                             print '----'
                            
                            Plaintiffs_vehicle.append(v)
                            continue
                        
                        else:
                        #TT:如果找到[原告]、[被告]&[交通工具]，但順序不是[被告]->[原告]->[交通工具]，則不管怎樣該交通工具就是被告的。
                        #TT:那如果[原告]->[交通工具]->[被告]這個順序在哪處理??
#                             print u'被告',v,c
#                             print ls[i]
#                             print '----'
                            
                            Defendants_vehicle.append(v)
                            continue
            
                    elif u'原告' in ls[i]:#TT:如果只找到原告在 ls[i]中，交通工具是原告的
#                         print u'原告',v,c
#                         print ls[i]
#                         print '----'
                        
                        Plaintiffs_vehicle.append(v)
                        continue

                    elif u'被告' in ls[i]:#TT:如果只找到被告在 ls[i]中，交通工具是被告的
#                         print u'被告',v,c
#                         print ls[i]
#                         print '----'
                        
                        Defendants_vehicle.append(v)
                        continue
     
    Plaintiffs_vehicle = list(set(Plaintiffs_vehicle))#TT:用set消掉重複的key name，做成list←WHY???
    Defendants_vehicle = list(set(Defendants_vehicle))
    plv = ','.join(Plaintiffs_vehicle) if len(Plaintiffs_vehicle) != 0 else 0 #TT:如果找到的交通工具種類不為一種，用逗號隔開串成一句。
    dev = ','.join(Defendants_vehicle) if len(Defendants_vehicle) != 0 else 0
    return plv,dev #有return所以就會有個地方接它?在哪???@@
print 'def find_vehicle(content) set'
```
